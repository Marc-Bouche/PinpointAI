```python
import requests
import random
import time
import pandas as pd  # Import pandas to create a DataFrame
import geopandas as gpd  # Import geopandas for geospatial data processing

# Set up API Keys
STREET_VIEW_API_KEY = "AIzaSyDugvJ210sXsrpBIRtR7B2tIsNNcFiJjqk"  # Replace with your Google Street View API Key
num_coordinates = 4000  # Number of coordinates to generate

# Load the France boundary from the JSON file (GeoJSON format)
france_gdf = gpd.read_file('/Users/mbouch17/Desktop/Personal_Data_Project/PinpointAI/france.json')  # Corrected path

# Function to generate random coordinates within the bounding box of France
def generate_random_coordinate():
    # Get the bounding box of the France polygon
    minx, miny, maxx, maxy = france_gdf.bounds.values[0]
    lat = random.uniform(miny, maxy)
    lon = random.uniform(minx, maxx)
    return lat, lon

# Function to check if a location has Google Street View coverage
def check_streetview_coverage(lat, lon):
    url = f"https://maps.googleapis.com/maps/api/streetview/metadata?location={lat},{lon}&key={STREET_VIEW_API_KEY}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        print(f"Response for {lat}, {lon}: {data['status']}")  # Debug print
        return data.get("status") == "OK"
    print(f"Failed request for {lat}, {lon}, status code: {response.status_code}")  # Debug if request fails
    return False

# List to store valid coordinates with Street View coverage
valid_coordinates = []

# Loop until we collect 5000 valid coordinates (or more)
while len(valid_coordinates) < num_coordinates:
    lat, lon = generate_random_coordinate()
    print(f"Generated coordinate: ({lat}, {lon})")  # Debug print for each generated coordinate
    if check_streetview_coverage(lat, lon):
        valid_coordinates.append((lat, lon))
        print(f"Added valid coordinate: ({lat}, {lon})")

    # Sleep to prevent hitting API rate limits (adjust as needed)
    time.sleep(0.1)  # Reduce sleep time if your API allows higher frequency calls

# After the loop, valid_coordinates will contain the valid coordinates with Street View coverage
print(f"Successfully collected {len(valid_coordinates)} coordinates with Street View coverage.")

# Store the coordinates in a pandas DataFrame
df_coordinates_fr5 = pd.DataFrame(valid_coordinates, columns=["Latitude", "Longitude"])

# Optionally, save the DataFrame to a CSV file
df_coordinates_fr5.to_csv("valid_coordinates_fr5.csv", index=False)

# Print the first few rows of the DataFrame for verification
print(df_coordinates_fr5.head())
```
